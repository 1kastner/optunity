We will use OpenCV (http://www.opencv.org/) for optical character recognition (OCR) using support vector machine (SVM) classifiers. This example is based on OpenCV's digit tutorial (available in `OPENCV_ROOT/samples/python2/digits.py`) and the following tutorial: http://opencv-python-tutroals.readthedocs.org/en/latest/py_tutorials/py_ml/py_svm/py_svm_opencv/py_svm_opencv.html#svm-opencv.

The feature vectors will constitute a histogram of gradients (HOG). More information is available at http://en.wikipedia.org/wiki/Histogram_of_oriented_gradients.

We start with the necessary imports.

```python
import cv2
import numpy as np
import optunity
import optunity.metrics
%matplotlib inline
import matplotlib.pyplot as plt

# finally, we import some convenience functions to avoid bloating this tutorial
# these are available in the `OPTUNITY/notebooks` folder
import opencv_utility_functions as util
```

We start by loading the data, which is available in the `OPTUNITY/notebooks` directory. This sample file is copied from OpenCV's example.

```python
digits, labels = util.load_digits("opencv-digits.png")
```

We preprocess the data and make a train/test split.

```python
rand = np.random.RandomState(321)
shuffle = rand.permutation(len(digits))
digits, labels = digits[shuffle], labels[shuffle]

digits2 = map(util.deskew, digits)
samples = util.preprocess_hog(digits2)

train_n = int(0.9*len(samples))

digits_train, digits_test = np.split(digits2, [train_n])
samples_train, samples_test = np.split(samples, [train_n])
labels_train, labels_test = np.split(labels, [train_n])

test_set_img = util.mosaic(25, digits[train_n:])
plt.title('Test set digits')
plt.imshow(test_set_img)
```

Now, it's time to construct classifiers. We will use a k-NN and SVM classifier. 

Lets start with the k-NN model, which has one hyperparameter $k$ that controls the number of neighbours to consider. The default value is $k=3$.

```python
knn_default = util.KNearest()
knn_default.train(samples_train, labels_train)
vis = util.evaluate_model(knn_default, digits_test, samples_test, labels_test)
plt.title('unoptimized k-NN test set performance')
plt.imshow(vis)
```

We will use a SVM classifier with an RBF kernel, i.e. $\kappa(\mathbf{u},\mathbf{v}) = \exp(-\gamma\|\mathbf{u}-\mathbf{v}\|^2)$. Such an SVM has two hyperparameters that must be optimized, namely the misclassification penalty $C$ and kernel parameter $\gamma$. First we define the objective function. We will use 5-fold cross-validated area under the ROC curve.

We start with an SVM with default parameters, which in OpenCV means: $C=1$ and $\gamma=0.5$.

```python
model = util.SVM()
model.train(samples_train, labels_train)
vis = util.evaluate_model(model, digits_test, samples_test, labels_test)
plt.title('unoptimized SVM test set performance')
plt.imshow(vis)
```

Next, we will construct a model with optimized hyperparameters. First we need to build Optunity's objective function. We will use 5-fold cross-validated error rate as loss function, which we will minimize.

```python
@optunity.cross_validated(x=samples_train, y=labels_train, num_folds=5)
def svm_error_rate(x_train, y_train, x_test, y_test, C, gamma):
    model = util.SVM(C=C, gamma=gamma)
    model.train(x_train, y_train)
    resp = model.predict(x_test)
    error_rate = (y_test != resp).mean()
    return error_rate
```

We will use Optunity's default solver to optimize $\gamma$ and $C$ with up to 100 function evaluations.

```python
optimal_parameters, details, _ = optunity.minimize(svm_error_rate, num_evals=100,
                                                   C=[0, 5], gamma=[0, 10])
print("Optimal parameters: C=%1.3f, gamma=%1.3f" % (optimal_parameters['C'], optimal_parameters['gamma']))
print("Cross-validated error rate: %1.3f" % details.optimum)
```

Finally, we train a model with the optimized parameters and determine its test set performance.

```python
model_opt = util.SVM(**optimal_parameters)
model_opt.train(samples_train, labels_train)
vis_opt = util.evaluate_model(model_opt, digits_test, samples_test, labels_test)
plt.title('optimized SVM test set performance')
plt.imshow(vis_opt)
```
